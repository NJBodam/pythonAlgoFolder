```python

def date_time_converter(time):
    
    
    year_in_min = 518400
    month_in_min = 43200
    day_in_min = 1440
    hour_in_min = 60
    
    
    year = time // year_in_min
    
    month = (time % year_in_min) // month_in_min
    
    day = ((time % year_in_min) % month_in_min) // day_in_min
    
    hour = (((time % year_in_min) % month_in_min) % day_in_min) // hour_in_min

    remaining = time % 60
    
    
    print(f"{year} years, {month} months, {day} days, {hour} hours, and {remaining} minutes.")

date_time_converter(1049190255)
```

```python
# Sensor Data App V1
# Jerry, Bodam
# Ramirez Rojas, Rocio




# Sensor Info 
sensor = input("Please enter the name of the sensor: ")
unit = input("Please enter the unit of the sensor’s measurements: ")
time_step = int(input("Please enter the time step of the measurements (in minutes): "))

print(f"\nThe time span between measurements is {time_step // (24 * 60)} days {time_step // 60 % 24} hours {time_step % 60} minutes.\n")

# First measurement
allValues = []  # Initializing an empty list

while True:
    
    if not allValues:
        date_str = input("Please enter the date of the first measurement (format yyyy-mm-dd): ")
        time_str = input("Please enter the time of the first measurement (format hh:mm): ")
    else:
        # calculate the next measurement
        next_measurement = allValues[-1]
        year, month, day, hour, minute = next_measurement[3:8]

        # convert current date and time
        current_total_minutes = day * 24 * 60 + hour * 60 + minute


        next_total_minutes = current_total_minutes + time_step

        # Extract next date and time
        day = next_total_minutes // (24 * 60)
        hour = (next_total_minutes % (24 * 60)) // 60
        minute = (next_total_minutes % (24 * 60)) % 60

        # next day is more than one month
        if day > 30:
            month += day // 30
            day %= 30

        # next month is more than one year
        if month > 12:
            year += month // 12
            month %= 12

        date_str = f"{year:04d}-{month:02d}-{day:02d}"
        time_str = f"{hour:02d}:{minute:02d}"

    # split date and time strings
    year, month, day = [int(part) for part in date_str.split('-')]
    hour, minute = [int(part) for part in time_str.split(':')]

    # value from the user
    value = float(input("Please enter the measured value (float): "))

    # tuple
    curValue = (sensor, unit, time_step, year, month, day, hour, minute, value)

    
    allValues.append(curValue)

    # print the current measurement
    print(f"\nSensor {sensor} measured {value} {unit} on {day}.{month}.{year} at {hour}:{minute}.")

    # another value?
    next_value = input("Do you want to enter another value? (y or n): ")
    if next_value.lower() != 'y':
        break

# print all values
print("\nAll Measurements:")
for value in allValues:
    print(value)


```

```python
# Team_C- 1 (2/3)
# Names: Jerry Bodam; Ramirez Rojas, Rocio
# Date: 24/11/2023
# Description: This program checks and count common letters between two words and letters in the same position.
# Warning: Some letters may be counted twice if they are repeated in the same word, or given the word with the minimum range.
# Professor said its ok.


word1 = input("Please enter a word: ").lower()
word2 = input("Please enter another word: ").lower()

# variables for common letters and thier count
common_letters_count = 0
common_letters = ""

# variables for letters in the same position and their count
same_position_count = 0
same_position_letters = ""

# Iterate over the range of the minimum length of the two words to avoid index out of range error
for i in range(min(len(word1), len(word2))):
    # Checks if the letters in the same position are the same and adds to the count and string if True
    if word1[i] == word2[i]:
        same_position_count += 1
        same_position_letters += word1[i]
    # Checks if the letter in word1 is in word2 and adds to the count and string if True
    if word1[i] in word2:
        common_letters_count += 1
        common_letters += word1[i]

print(f"'{word1}' and '{word2}' have {common_letters_count} common letters: '{common_letters}'.")
print(f"'{word1}' and '{word2}' have {same_position_count} letters at the same position: '{same_position_letters}'.")
```
